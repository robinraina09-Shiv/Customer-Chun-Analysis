```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\user\Customer-Churn-Analysis\Dataset\Telco-Customer-Churn.csv")
```

```python
contract_churn = pd.crosstab(df["Contract"], df["Churn"])

contract_churn.plot(kind="bar", figsize=(8,5))

plt.title("Churn by Contract Type")
plt.xlabel("Contract")
plt.ylabel("Customers")
plt.xticks(rotation=10)

plt.show()
```

```python
Internet_Churn = pd.crosstab(df["InternetService"], df["Churn"])

Internet_Churn.plot(kind="bar", figsize=(8,5))

plt.title("Churn by Internet Service")
plt.xlabel("Internet Service")
plt.ylabel("Customers")
plt.xticks(rotation=10)

plt.show()
```

```python
Payment_Churn = pd.crosstab(df['PaymentMethod'], df['Churn'])
Payment_Churn.plot(kind='bar', figsize=(8,5))

plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Customers")
plt.xticks(rotation=10)

plt.show()
```

```python
Gender_Churn = pd.crosstab(df['gender'], df['Churn'])   
Gender_Churn.plot(kind='bar', figsize=(8,5))    
plt.title("Churn by Gender")
plt.xlabel("Gender")    
plt.ylabel("Customers")
plt.xticks(rotation=10)
plt.show()
```

```python
SeniorCitizen_Churn = pd.crosstab(df['SeniorCitizen'], df['Churn'])
SeniorCitizen_Churn.plot(kind='bar', figsize=(8,5))
plt.title("Churn by Senior Citizen Status")
plt.xlabel("Senior Citizen")
plt.ylabel("Customers")
plt.xticks(rotation=10)
plt.show()
```

```python
plt.Figure(figsize=(8,5))
df.boxplot(column='MonthlyCharges', by='Churn')
plt.title("MonthlyCharges by Churn")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("MonthlyCharges")
plt.show()
```

```python
plt.figure(figsize=(7,5))

df.boxplot(column="TotalCharges", by="Churn")

plt.title("Total Charges by Churn")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Total Charges")

plt.show()
```

```python
(df["TotalCharges"] == " ").sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

print(df["TotalCharges"].dtype)
```

```python
numeric_df = df.select_dtypes(include=["number"])

corr = numeric_df.corr()

plt.figure(figsize=(6,5))
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
