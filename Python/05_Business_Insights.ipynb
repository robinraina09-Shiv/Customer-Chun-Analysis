                         Convert Month-to-Month Customers

Observation - 
              Month-to-month customers have the highest churn rate (42.71%).

Recommendation - 
              Offer discounts or loyalty incentives to encourage customers to switch to one-year or two-year contracts.

Expected Business Impact - 
              Lower churn and improve customer retention.

                         Improve Fiber Optic Customer Experience

Observation - 
            Fiber optic customers have a churn rate of 41.89%, the highest among internet service types.

Recommendation - 
            Investigate pricing, service quality, and customer support issues.
            Conduct customer feedback surveys.

Expected Business Impact - 
            Increase customer satisfaction and reduce churn.

                            Promote Automatic Payments

Observation - 
             Customers using Electronic Check have a churn rate of 45.29%.

Recommendation - 
             Encourage customers to switch to Credit Card or Bank Transfer (automatic) using cashback or discounts.

Expected Business Impact - 
             Reduce payment-related friction and improve retention.

                              Increase Adoption of Tech Support

Observation - 
             Customers without Tech Support have a churn rate of 41.64%.

Recommendation - 
             Bundle Tech Support with premium plans or offer a free trial.

Expected Business Impact - 
             Increase customer engagement and reduce churn.

                             Increase Adoption of Online Security

Observation - 
             Customers without Online Security have a churn rate of 41.77%.

Recommendation - 
             Bundle Online Security with internet plans.

Expected Business Impact - 
             Improve customer loyalty and perceived value.

                             Focus on Senior Citizens

Observation - 
               Senior Citizens have a churn rate of 41.68%.

Recommendation - 
               Provide dedicated customer support and targeted loyalty programs.

Expected Business Impact - 
                        Improve retention among a high-risk customer segment.

                          Executive Summary
Key Findings
Overall churn rate: 26.54%

Revenue lost due to churn: $2,862,926.90

Month-to-month customers are the highest-risk segment.

Fiber optic customers churn more than DSL customers.

Electronic Check users have the highest churn.

Tech Support and Online Security significantly reduce churn.

Senior Citizens have a higher churn rate than other customers.
